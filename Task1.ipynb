Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('worldpopulationdata.csv')
```

```python
print(data.head())
```

```python
print(data.describe())
```

```python
# List of countries for the graph
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

# Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]
```

```python
# Creating the bar chart
plt.figure(figsize=(10, 6))  # Set figure size for better readability
plt.bar(filtered_data['Country Name'], filtered_data['2022'], color='skyblue', edgecolor='black')
plt.xlabel('Country Name')
plt.ylabel('Population in 2022')
plt.title('Population of selected countries in 2022')
```
